# Phase-2 -  Model (Prof. Given Answers) and Similarity scores Implemtation

we first experimented with several text encoding techniques, including TF-IDF and Jaccard similarity. However, we encountered several drawbacks with these methods. With TF-IDF, the length of the comparing vectors had to match, which was not suitable for our case as student answers could be of any length. Therefore, we dropped this method. With Jaccard similarity, the similarity score only matched each word and not the context, which was also not appropriate for our purposes. Thus, we dropped this method as well.
Finally, we decided to use the BERT (Bidirectional Encoder Representations from Transformers) model to encode the text, as it can encode sequences of varying lengths and is able to capture the context of the text. We fine-tuned the BERT model on our training data, which consisted of both the student's answers and the professor's correct answers. This allowed us to obtain an encoding for each set of answers.
Our methodology involved encoding the test using the BERT model and then using cosine similarity to calculate the similarity score between the student's answers and the professor's correct answers. This allowed us to develop an accurate and efficient automatic grading system for student answers

## Jaccard Similarity

---



```python
from sklearn.metrics import jaccard_score

# Define a function to calculate Jaccard similarity score
def jaccard_similarity(x, y):
    intersection_cardinality = len(set.intersection(*[set(x.split()), set(y.split())]))
    union_cardinality = len(set.union(*[set(x.split()), set(y.split())]))
    return intersection_cardinality / float(union_cardinality)

# Calculate Jaccard similarity score and add it as a new column
df['Jaccard Similarity'] = df.apply(lambda row: jaccard_similarity(row['answers'], row['all_answers']), axis=1)
```
